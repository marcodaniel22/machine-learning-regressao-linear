```python
# imports de bibliotecas
import pandas as pd
import numpy as np
from numpy import nan as NA
```

```python
# Definiçao de caminhos e arquivos
basePath = "./dados/COMBUSTIVEL/"
filename = "SP_DATA.csv"
produtos = ["DIESEL","ETANOL", "GASOLINA", "GNV"]
```

```python
# Leitura de dataset
data = pd.read_csv(basePath + filename)
df = pd.DataFrame(data)
```

```python
# Para cada bloco de dados por produto:
#    é abstraido a data em formata Ano/Mes
#    o preço de venda dos combustives é comvertido para float
dataframes = []
for prod in produtos:
    df_prod = df[df["Produto"] == prod]
    novo_df_prod = pd.DataFrame()
    novo_df_prod["DATA"] = df_prod["Data"].apply(lambda x: x.split("/")[2] + "/" + x.split("/")[1])
    novo_df_prod[prod] = df_prod["Venda"].astype(float)
    novo_df_prod = novo_df_prod.groupby("DATA",as_index =False).mean()
    dataframes.append(novo_df_prod)
```

```python
# Salva arquivo csv dos dados dos combustiveis unificados separados por coluna
comb_df = pd.concat(dataframes, sort=False)
comb_df = comb_df.groupby("DATA",as_index =False).mean()
comb_df
```

```python
comb_df.to_csv(basePath + "SP_COMB_DATA.csv", sep=",", index=False)
```
