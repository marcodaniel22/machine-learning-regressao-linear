```python
# imports de bibliotecas
import pandas as pd
import numpy as np
```

```python
# Definiçao de caminhos e arquivos
basePath = "./dados/DOLAR/"
filename = "USD_BRL.csv"
```

```python
# Leitura do arquivo baixado do Yahoo Finanças
data = pd.read_csv(basePath + filename)
df = pd.DataFrame(data)
df.head(12)
```

```python
# Funçao para converter strings para float
def convertToFloat(value):
    replaced = value.replace(",", ".")
    return float(replaced)

# Funçao para obter a media dos campos de interesse
def getMeanValues(df_values):
    values = []
    for row in df_values:
        values.append(row.mean())
    return values

# Funçao que trata a data do registro para o formato Ano/Mes
meses = {"Dez":"12", "Nov":"11", "Out":"10", "Set":"09","Ago":"08", "Jul":"07", 
         "Jun":"06", "Mai":"05", "Abr":"04", "Mar":"03", "Fev":"02", "Jan":"01"}
def getFormatDate(date):
    splited = date.split(" ")
    ano = "20" + str(splited[1])
    return ano + "/" + meses[splited[0]]
```

```python
# Conversao dos dados para float
new_df = pd.DataFrame()
new_df["Último"] = df["Último"].apply(convertToFloat)
new_df["Abertura"] = df["Abertura"].apply(convertToFloat)
new_df["Máxima"] = df["Máxima"].apply(convertToFloat)
new_df["Mínima"] = df["Mínima"].apply(convertToFloat)

# Uniao dos valores por meio de média aritimetica
final_df = pd.DataFrame()
final_df["DATA"] = df["Data"].apply(getFormatDate)
final_df["DOLAR"] = getMeanValues(new_df[["Último", "Abertura", "Máxima", "Mínima"]].values)
final_df = final_df[::-1]

# Salva arquivo CSV dos dados tratados
final_df.to_csv(basePath + "DOLAR_DATA.csv", sep=",", index=False)
```
